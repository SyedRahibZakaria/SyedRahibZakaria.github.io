```
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Importing dataset
dataset = pd.read_csv('Marketing Raw Data.csv')
```

```
dataset.head()
```

```
# Line Graph

ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', data = dataset)

# Note: The error bands (light blue region) show the 95% confidence interval
```

```
# Adding categories
ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', data = dataset, hue = 'Promo')
```

```
# Changing style with respect to Promo categories
ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', data = dataset, hue = 'Promo', style='Promo')
```

```
# Different style wrt. Promo, increase size and remove confidence interval

sns.set(rc={'figure.figsize': (10,8)})
ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', data = dataset, hue = 'Promo', style='Promo', ci = False)
```

```
# Adding markers
ax = sns.lineplot(x = 'Month_ID', y = 'Revenue', data = dataset, hue = 'Promo', style='Promo', ci = False, markers = True)
```

```
# By Day_Name with style and adding markers by month

ax = sns.lineplot(x = 'Month_ID', y = 'Revenue', data = dataset, hue = 'Day_Name', style='Promo', ci = False, markers = True)
```

```
# By Day_Name with style and adding markers by month

ax = sns.lineplot(x = 'Year', y = 'Revenue', data = dataset, hue = 'Day_Name', style='Promo', ci = False, markers = True)
```

Bar Plot

```
# Total Revenue by Month
ax = sns.barplot(x = 'Month_ID', y = 'Revenue', data = dataset)

# Notes:

#1. The lines signify the confidence interval
#2. Takes mean of the revenue and not the sum of total revenue in each month 
```

```
dataset[['Revenue', 'Month_ID']].groupby('Month_ID').agg({'Revenue':'mean'})
```

```
# total Revenue by month - remove confidence interval - By Promo
ax = sns.barplot(x = 'Month_ID', y = 'Revenue', data = dataset, ci = None, hue = 'Promo')
```

```
# Flipping bar chart horizontally

ax = sns.barplot(x = 'Revenue', y = 'Month_ID', data = dataset, ci = None, hue = 'Promo', orient = 'h')
```

```
# Adding color

ax = sns.barplot(x = 'Revenue', y = 'Month_ID', data = dataset, ci = None, hue = 'Promo', orient = 'h', color = '#1CA7C1')

# Pick color from website
# https://htmlcolorcodes.com/color-picker/
```

**Histograms**

```
# Useful to investigate the distribution of data

# Investigating the distribution of Revenue
# 1st we need to specify an array that has only Revenue

x = dataset['Revenue'].values

ax = sns.distplot(x, color = '#239DE0')
```

```
# Investigating the distribution of Revenue, adding the mean using matplotlib ("axvline" command)


# Calculating mean
mean = dataset['Revenue'].mean()
sns.distplot(x, color = '#E02373')  # #2370E0

# Plotting the mean using matplotlib
plt.axvline(mean, 0, 1, color = '#9423E0') #
```

Box Plots

```
# Investigating Distribution

x = dataset['Revenue'].values

sns.set(rc = {'figure.figsize': (10,8)})
sns.boxplot(x)

print(dataset['Revenue'].median())
# Notes:

# 1. The line in the middle of the box signifies the median
# 2. The box in the middle show the begining of Q1 (25th percentile) and the end of Q3 (75th percentile)
# 3. The whiskers (left - right) show the minimum quartile and maximum quartile
# 4. The dots on the right are "outliers"
# more details: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51
```

```
# Investigating the distribution of Revenue by Day

ax = sns.boxplot(x = 'Day_Name', y = 'Revenue', data = dataset)
```

```
# Investigating distribution of Revenue - Flip Horizontally - Change color

ax = sns.boxplot(x = 'Revenue', y = 'Day_Name', data = dataset, color = '#F18C20')
```

```
# Adding hue

ax = sns.boxplot(x = 'Day_Name', y = 'Revenue', data = dataset, color = '#F18C20', hue = 'Promo')
```

```
# Investigating the distribution of Revenue by Day - by color - by data points(swarmplot)

ax = sns.boxplot(x = 'Day_Name', y = 'Revenue', data = dataset, color = '#1EC3F1')
ax = sns.swarmplot(x = 'Day_Name', y = 'Revenue', data = dataset, color = '#F18C20')
```
